```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from arch import arch_model
from statsmodels.graphics.tsaplots import plot_acf
from statsmodels.stats.diagnostic import acorr_ljungbox
from scipy.stats import t
import pandas_datareader.data as web
```

## Step 1: Load Data

```python
# Get Los Angeles Temperature
data = pd.read_csv(r"C:\Users\sarah\Desktop\BC\Fall 2025\MESA8414 Applied AI and Machine Learning\Assignment\Week 4 assignment\temperature.csv", usecols=["datetime", "Los Angeles"])

# Compute log changes
data['Change'] = np.log(data['Los Angeles'] / data['Los Angeles'].shift(1)) * 100 
returns = data['Change'].dropna()

# Plot ###FILL IN BLANKS BELOW
plt.figure(figsize=(12, 6))
plt.plot(returns)
plt.title('Log Changes in Los Angeles Temperature')
plt.xlabel('Date')
plt.ylabel('Change (%)')
plt.grid(True)
plt.show()
```

## Step 2: Fit GARCH(1,1) Model

```python
model1 = arch_model(returns, vol='Garch', p=1, q=1, mean='Constant', dist='normal')
res1 = model1.fit(disp='off')
print(res1.summary()) ###FILL IN THE BLANK TO PRINT THE SUMMARY
```

## Step 3: Fit EGARCH(1,1) Model

```python
model2 = arch_model(returns, vol='EGARCH', p=1, q=1, mean='Constant', dist='normal')
res2 = model2.fit(disp='off')
print(res2.summary())
```

## Step 4: Compare AIC/BIC

```python
print("GARCH AIC:", res1.aic, "| BIC:", res1.bic)
print("EGARCH AIC:", res2.aic, "| BIC:", res2.bic)
```


What does the above analysis tell us with respect to model fit and complevity?

By comparing AIC and BIC across the two models, we can learn about which model is comparatively a better fitting and more parsimonious model. In this case, GARCH(1,1) model is the better fitting and more parsimonious model. 

## Step 5: Residual Diagnostics - ACF

```python
plot_acf(res2.std_resid) ###FILL IN THE BLANK
plt.title("ACF of Standardized Residuals (EGARCH)")
plt.show()
```

Residual diagnostics are important to assess whether the EGARCH model has properly captured the dynamics in the return series, particularly autocorrelation and time-varying volatility. Interpret the above plot.

The plot of the autocorrelation function (ACF) of the standardized residuals shows cyclical patterns of autocorrelation. This indicates that the model did not fully remove the noise and cyclical patterns present in the dataset. 

## Step 5: Ljung-Box Test

```python
lb_test = acorr_ljungbox(res2.std_resid, lags=24, return_df=True)
print(lb_test)
```

The LjungBox test is used to detect autocorrelation in the residuals of a time series model. It tests the null hypothesis that the residuals are independently distributed up to a certain lag (in this case, lag 10). Interpret the results.

The Ljung-Box test results returned p-values that are all below .001. Therefore, the test rejects the null hypothesis that residuals are independently distributed. This indicates that the residuals may be autocorrelated. 

## Step 7: Rolling Forecast (Expanding Window)

```python
rolling_preds = []
true_vals = []
window = 1000

scaled_returns = returns *10 
for i in range(window, len(scaled_returns) - 1):
    train = scaled_returns.iloc[:i]
    test = scaled_returns.iloc[i + 1]

    model = arch_model(train, vol='EGARCH', p=1, q=1, mean='AR', lags=1, dist='t') ###FILL IN THE BLANKS
    result = model.fit(disp='off')
    forecast = result.forecast(horizon=1)
    sigma = np.sqrt(forecast.variance.values[-1, 0])
    rolling_preds.append(sigma)
    true_vals.append(abs(test))
    

plt.figure(figsize=(12, 4))
plt.plot(returns.index[window + 1:], true_vals, label="|Actual Return|")
plt.plot(returns.index[window + 1:], rolling_preds, label="Forecast Volatility")
plt.legend()
plt.title("Rolling Forecast: EGARCH(1,1)")
plt.show()
```

## Step 8: Dynamic Value at Risk (VaR)

```python
from scipy.stats import t

alpha = 0.05
df = res2.params['nu']
t_quant = t.ppf(alpha, df=df)

VaR = res2.params['Const'] + res2.conditional_volatility* t_quant ### ADD THE OPERATIONS TO THE LINE

plt.figure(figsize=(12, 4))
plt.plot(returns.index, returns, label="Returns")
plt.plot(returns.index, VaR, label="VaR (95%)", color='red')
plt.title("Dynamic VaR using EGARCH Model (95% Confidence)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

Note: In the financial world, Value at Risk (VaR) is a widely used risk metric that estimates the potential loss in value of a portfolio or asset over a given time horizon, at a specific confidence level.
