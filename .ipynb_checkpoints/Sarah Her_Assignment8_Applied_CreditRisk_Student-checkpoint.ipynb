### Theoretical Background  

#### 1 Support Vector Machines  
SVM is a supervised learning algorithm that finds the optimal hyperplane that best separates the data into classes.  
It maximizes the margin between the nearest data points of different classes (called support vectors).  

**Mathematical Formulation:**  
For a binary classification task, SVM solves:  

$$
\min_{w,b}\frac{1}{2}\|w\|^2 \quad
\text{s.t. } y_i(w^Tx_i+b)\ge1
$$  

The decision function is:  
$$
f(x)=\text{sign}(w^Tx+b)
$$  

---

#### 2 Kernel Trick  
Real-world data is often non-linearly separable.  
SVM handles this by using **kernel functions** to map data into a higher-dimensional space where a linear separator can be found.  

Common kernels:  
- Linear Kernel: \( K(x_i,x_j)=x_i^Tx_j \)  
- Polynomial Kernel: \( K(x_i,x_j)=(x_i^Tx_j+c)^d \)  
- RBF Kernel: \( K(x_i,x_j)=\exp(-\gamma \|x_i-x_j\|^2) \)  

**Linear SVM** works well for high-dimensional text or numeric features.  
**RBF SVM** captures complex, non-linear patterns in the data.  

---

#### 3 Regularization and Hyperparameters  
- **C (parameter):** controls the trade-off between margin width and classification errors.  
  - Low C → wider margin, more tolerance for errors (high bias).  
  - High C → narrow margin, less tolerance for errors (low bias).  
- **Gamma (for RBF):** defines how far the influence of a single training example reaches.  
  - Low gamma → smooth decision boundary.  
  - High gamma → tight fit to training data (overfitting risk).  

---

#### 4 Evaluation Metrics  

| Metric | Formula | Interpretation |  
|:--|:--|:--|  
| Accuracy | \( \frac{TP+TN}{TP+TN+FP+FN} \) | Overall correctness |  
| Precision | \( \frac{TP}{TP+FP} \) | Among predicted defaults, how many were true defaults |  
| Recall | \( \frac{TP}{TP+FN} \) | Proportion of actual defaults correctly identified |  
| F1 Score | \( 2 · \frac{Precision·Recall}{Precision+Recall} \) | Balance between precision and recall |  
| ROC-AUC | Area under ROC curve | Measures class separability |  

---

### Business Context

# Step 1 – Import Libraries and Load Data

```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.metrics import classification_report, confusion_matrix, precision_recall_curve, average_precision_score

# Load dataset
df = pd.read_csv(r'C:\Users\sarah\Desktop\BC\Fall 2025\MESA8414 Applied AI and Machine Learning\Assignment\Week 8 assignment\creditcard-2.csv')   ### FILL IN BLANK

# Display dataset information
print("Dataset shape:", df.shape)  
print("\nColumns:\n", df.columns)  
print("\nData Info:\n")
print(df.info())

# Display first few rows
df.head()

```

```python
# Visualize class distribution
print("Class distribution:")
print(df['Class'].value_counts())   ### FILL IN BLANK

sns.countplot(data=df, x='Class')   ### FILL IN BLANK
plt.title('Class Distribution (0 = Genuine, 1 = Fraud)')
plt.show()
```

# Step 2: Sampling for Computational Efficiency  


```python
# Perform random sampling for efficiency
sample_df = df.sample(n=10000, random_state=21)   ### FILL IN BLANK
print("Sampled Data Shape:", sample_df.shape)

# Check class balance in the sampled data
print(sample_df['Class'].value_counts(normalize=True))   ### FILL IN BLANK

```

# Step 3: Feature Scaling and Train-Test Split

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

X = sample_df.drop("Class", axis=1)   ### FILL IN BLANK
y = sample_df["Class"]                ### FILL IN BLANK

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)   ### FILL IN BLANK

X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42, stratify=y   ### FILL IN BLANKS
)

print("Training samples:", X_train.shape[0], " Testing samples:", X_test.shape[0])
```

**Interpretation Question:** What might happen if we skip scaling when using distance-based algorithms like SVM?

* When scaling is not used in distance-based algorithms like SVM, variables with a large numeric range could dominate the distance computation. For example, the credit card balance will tend to be much higher in number when compared to the number of years having that credit card. In this case, without scaling, SVM could pay attention mostly to the differences in the credit card.

# Step 4: Baseline Linear SVM
Using a **linear kernel** with class balancing enabled.


```python
from sklearn.svm import SVC
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score

linear_svm = SVC(kernel='linear', class_weight='balanced', probability=True, random_state=42)   ### FILL IN BLANKS
linear_svm.fit(X_train, y_train)   ### FILL IN BLANK
y_pred_linear = linear_svm.predict(X_test)   ### FILL IN BLANK

print("\n--- Linear Kernel SVM ---")
print(classification_report(y_test, y_pred_linear, digits=4))   ### FILL IN BLANK
print("ROC-AUC Score:", roc_auc_score(y_test, linear_svm.predict_proba(X_test)[:,1]))   ### FILL IN BLANKS

cm_linear = confusion_matrix(y_test, y_pred_linear)   ### FILL IN BLANK
print("\nConfusion Matrix:\n", cm_linear)
```

**Interpretation Question:** Why might a linear boundary perform poorly on complex, nonlinear data?

* A linear boundary will perform poorly on complex, nonlinear data because having a linear boundary means that you're dividing classes by a straight line. Therefore, if the data is nonlinear, meaning that it has some circular or curved shape, a straight line will not effectively separate classes that fit the data. 

# Step 5: Faster RBF SVM (No Grid Search)
Train an **RBF kernel SVM**, which is capable of handling non-linear relationships.


```python
rbf_svm = SVC(kernel='rbf', C=1.0, gamma='scale', class_weight='balanced', probability=True, random_state=42)   ### FILL IN BLANKS
rbf_svm.fit(X_train, y_train)   ### FILL IN BLANK
y_pred_rbf = rbf_svm.predict(X_test)   ### FILL IN BLANK

print("\n--- RBF Kernel SVM ---")
print(classification_report(y_test, y_pred_rbf, digits=4))   ### FILL IN BLANK
print("ROC-AUC Score:", roc_auc_score(y_test, rbf_svm.predict_proba(X_test)[:,1]))   ### FILL IN BLANKS

cm_rbf = confusion_matrix(y_test, y_pred_rbf)   ### FILL IN BLANK
print("\nConfusion Matrix:\n", cm_rbf)

```

**Interpretation Question:** How does the RBF kernel differ from the linear kernel in handling nonlinear relationships?

* A linear kernel fits a straight line, but an RBF kernel could fit a flexible boundary that is nonlinear. 

# Step 6: Compare Model Performances
Evaluate and compare Linear vs RBF SVMs using key metrics.


```python
from sklearn.metrics import accuracy_score, f1_score

results = pd.DataFrame({
    "Model": ["Linear SVM", "RBF SVM"],
    "Accuracy": [
        accuracy_score(y_test, y_pred_linear),   ### FILL IN BLANK
        accuracy_score(y_test, y_pred_rbf)    ### FILL IN BLANK
    ],
    "F1 Score": [
        f1_score(y_test, y_pred_linear),         ### FILL IN BLANK
        f1_score(y_test, y_pred_rbf)          ### FILL IN BLANK
    ],
    "ROC-AUC": [
        roc_auc_score(y_test, linear_svm.predict_proba(X_test)[:,1]),   ### FILL IN BLANK
        roc_auc_score(y_test, rbf_svm.predict_proba(X_test)[:,1])       ### FILL IN BLANK
    ]
})

print("\n--- Model Performance Summary ---")
display(results)
```

**Interpretation Question:** Which model shows the highest accuracy or F1 score, and what does that imply?

* Linear SVM has the highest F1 score, and it implies that the data had a linear relationship where a straight-line boundary fit the data better.

# Step 7: Visualize ROC Curves


```python
y_scores = rbf_svm.decision_function(X_test)   ### FILL IN BLANK
precision, recall, _ = precision_recall_curve(y_test, y_scores)   ### FILL IN BLANK
average_precision = average_precision_score(y_test, y_scores)     ### FILL IN BLANK

plt.figure(figsize=(8,6))
plt.plot(recall, precision, label=f'RBF SVM (AP={average_precision:.4f})')   ### FILL IN BLANK
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title('Precision-Recall Curve')
plt.legend()
plt.grid(True)
plt.show()

```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve, auc

fpr_lin, tpr_lin, _ = roc_curve(y_test, linear_svm.predict_proba(X_test)[:,1])   ### FILL IN BLANK
fpr_rbf, tpr_rbf, _ = roc_curve(y_test, rbf_svm.predict_proba(X_test)[:,1])   ### FILL IN BLANK

plt.figure(figsize=(8,6))
plt.plot(fpr_lin, tpr_lin, label=f"Linear SVM (AUC = {auc(fpr_lin, tpr_lin):.3f})")   ### FILL IN BLANK
plt.plot(fpr_rbf, tpr_rbf, label=f"RBF SVM (AUC = {auc(fpr_rbf, tpr_rbf):.3f})")     ### FILL IN BLANK
plt.plot([0,1],[0,1],'k--')
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve Comparison: Linear vs RBF SVM")
plt.legend()
plt.grid(True)
plt.show()
```
