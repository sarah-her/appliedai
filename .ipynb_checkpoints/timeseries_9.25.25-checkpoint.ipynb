## Step 7: Rolling Forecast (Expanding Window)

```python
rolling_preds = []
true_vals = []
window = 1000

for i in range(window, len(y) - 1):
    train = y.iloc[:i]
    test = y.iloc[i + 1]

    model = arch_model(train, vol='GARCH', p=1, q=1, mean='ARX', dist='t') 
    result = model.fit(disp='off')
    forecast = result.forecast(horizon=1)
    sigma = np.sqrt(forecast.variance.values[-1, 0])
    rolling_preds.append(sigma)
    true_vals.append(abs(test))
    

plt.figure(figsize=(12, 4))
plt.plot(y.index[window + 1:], true_vals, label="|Actual Return|")
plt.plot(y.index[window + 1:], rolling_preds, label="Forecast Volatility")
plt.legend()
plt.title("Rolling Forecast: GARCH(1,1)")
plt.show()
```

```python
from scipy.stats import norm

alpha = 0.05  
z_alpha = norm.ppf(alpha)

mu = pd.Series(res1.params['Const'], index=y.index)

sigma = res1.conditional_volatility.reindex(y.index)
VaR = mu + sigma * z_alpha

plt.figure(figsize=(12, 4))
plt.plot(y.index, y, label="Returns")
plt.plot(y.index, VaR, label="VaR (95%)", color='red')
plt.title("Dynamic VaR using GARCH Model (95% Confidence)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

## Step 9: Fit an ARIMA Model for Forecasting SPY Price

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA

# ----------------------------
# 1) Config
# ----------------------------
h = 100   # forecast horizon
spy_col = 'SPY Price'
btc_col = 'Close'  # Bitcoin price column (BTC)

# ----------------------------
# 2) Clean & align data
# ----------------------------
df = merged[['Date', spy_col, btc_col]].copy()
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
df = (df.dropna(subset=['Date'])
        .sort_values('Date')
        .drop_duplicates('Date')
        .set_index('Date')
        .asfreq('B'))  # business days

df[spy_col] = pd.to_numeric(df[spy_col], errors='coerce').ffill()
df[btc_col] = pd.to_numeric(df[btc_col], errors='coerce').ffill()

# avoid log(0)
eps = 1e-8
df[spy_col] = df[spy_col].clip(lower=eps)
df[btc_col] = df[btc_col].clip(lower=eps)

# ----------------------------
# 3) Endog & Exog
# ----------------------------
log_spy = np.log(df[spy_col])
btc_ret = np.log(df[btc_col]).diff()

aligned = pd.concat([log_spy.rename('log_spy'),
                     btc_ret.rename('btc_ret')], axis=1).dropna()
y = aligned['log_spy']
X = aligned[['btc_ret']]

# ----------------------------
# 4) Fit ARIMAX model
# ----------------------------
order = (1, 1, 1)
res = ARIMA(y, order=order, exog=X).fit()
print(res.summary())

# ----------------------------
# 5) Build future exogenous values
# ----------------------------
future_idx = pd.date_range(y.index[-1] + pd.offsets.BusinessDay(),
                           periods=h, freq='B')

# Placeholder: repeat last BTC return for future (replace with your BTC forecast if available)
last_btc_ret = float(X.iloc[-1, 0])
btc_ret_future = pd.Series(last_btc_ret, index=future_idx, name='btc_ret')
X_future = btc_ret_future.to_frame()

# ----------------------------
# 6) Forecast SPY log-price
# ----------------------------
fc = res.get_forecast(steps=h, exog=X_future)

log_mean = fc.predicted_mean
log_ci = fc.conf_int()

price_mean = np.exp(log_mean)
price_ci_lower = np.exp(log_ci.iloc[:, 0])
price_ci_upper = np.exp(log_ci.iloc[:, 1])

# ----------------------------
# 7) Build forecast table
# ----------------------------
forecast_table = pd.DataFrame({
    'Date': price_mean.index,
    'Forecast Price': price_mean.round(2),
    'Lower 95% CI': price_ci_lower.round(2),
    'Upper 95% CI': price_ci_upper.round(2)
}).reset_index(drop=True)

print(forecast_table.head(100))  
forecast_table.to_csv('spy_price_forecast.csv', index=False)
print("Saved 100-day forecast to spy_price_forecast.csv")

# ----------------------------
# 8) Plot observed + forecast
# ----------------------------
plt.figure(figsize=(14, 6))
plt.plot(df.index, df[spy_col], label='Observed SPY Price', color='black')
plt.plot(price_mean.index, price_mean, label='Forecast (100-day)', color='blue')
plt.fill_between(price_mean.index,
                 price_ci_lower,
                 price_ci_upper,
                 color='skyblue', alpha=0.3,
                 label='95% CI')
plt.title('SPY Price Forecast (ARIMAX with BTC exogenous)')
plt.xlabel('Date')
plt.ylabel('SPY Price')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
